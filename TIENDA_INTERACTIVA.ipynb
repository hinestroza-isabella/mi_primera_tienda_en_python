```python
# tienda interactiva
nombre = input("Hola! Bienvenidx ¿Cuál es tu nombre?\n")
deseo = input("¿Qué producto quieres comprar hoy? (hoodie, pantalon baggy, boxy tee, sneakers o una gorra 5 paneles)\n").lower()
precios = {
    "hoodie" : 200000,
    "pantalon baggy" : 250000,
    "boxy tee" : 150000,
    "sneakers" : 500000,
    "gorra 5 paneles" : 75000
}
# Selección de precio
if deseo in precios:
    precio_final = precios[deseo]
else:
    print("Lo siento, pero no tenemos este producto disponible")
    quit()
# Si el producto existe:
unidades = int(input(f"¿Cuántas unidades quieres, {nombre}?\n"))
total = unidades * precio_final
if unidades <= 0:
  print("Debes agregar mínimo una (1) unidad")
  quit()
# Verificación de dinero
dinero = int(input("¿Cuánto dinero tienes?\n"))
if dinero >= total:
    cambio = dinero - total
    print("Compra exitosa!") 
    print(f"Has comprado {deseo}, de manera exitosa")
    print(f"Nombre: {nombre}")
    print(f"Producto: {deseo}")
    print(f"Precio x unidad: ${precio_final:,}")
    print(f"Unidades: {unidades}")
    print(f"Total: ${total:,}")
    print(f"Tu cambio es de: ${cambio:,}")
    print("Gracias por tu compra, vuelve pronto!")
else:
    faltante = total - dinero
    print(f"Lastimosamente no te alcanza, te faltan ${faltante:,}, nos veremos en una próxima ocasión")
    
```
